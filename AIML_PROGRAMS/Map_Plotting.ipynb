```python
x = [5,2,7,5]
y = [2,16,4,2]
plt.plot(x,y)
plt.title('Info')
plt.ylabel('Y axis')
plt.xlabel('X axis')
plt.show()
```

```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y = [2, 3, 5, 7, 11]

font_0 = {'family': 'serif', 'style': 'italic'}

plt.plot(x, y, '--*', color='#70a431')
plt.xlabel('The x label')
plt.ylabel('The y label')
plt.title('A matplotlib plot', fontdict=font_0, fontsize=15, loc='left')

plt.show()
```
